```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
K = np.array([1,2,4,6,8,10])
fig, ax = plt.subplots(figsize=(10,10))

# Plot protonet results
protonet_acc = np.array([0.489, 0.596, 0.673, 0.714, 0.726, 0.737])
protonet_ci = np.array([0.008, 0.009, 0.008, 0.008, 0.008, 0.007])
ax.plot(K, protonet_acc, color='b', label='protonet')
ax.fill_between(K, (protonet_acc-protonet_ci), (protonet_acc+protonet_ci), color='b', alpha=.1)

# Plot MAML results
maml_acc = np.array([0.979, 0.990, 0.994, 0.993, 0.994, 0.995])
maml_ci = np.array([0.003, 0.002, 0.001, 0.002, 0.001, 0.001])
ax.plot(K, maml_acc, color='r', label='maml')
ax.fill_between(K, (maml_acc-maml_ci), (maml_acc+maml_ci), color='r', alpha=.1)

# Add labels and legend
ax.set_xlabel('K')
ax.set_ylabel('Test accuracy')
ax.legend()
```
